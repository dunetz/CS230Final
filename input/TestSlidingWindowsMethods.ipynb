```python
#https://stackoverflow.com/questions/41879440/how-to-implement-a-sliding-window-in-tensorflow
# NOTE: If you call this function often, you may want to (i) move the `np.pad()`
# into the graph as `tf.pad()`, and (ii) replace `X_t` with a placeholder.
sess=tf.Session()
window_size=5
X=np.arange(1000).reshape(10,100)
seqs=X.shape[-1]-window_size
X_t  = tf.constant(X,dtype=tf.float32)

def window_func(start):
    end = start + window_size
    X_matrix = tf.reshape(X_t[:,start:end],(-1,window_size))
    return X_matrix
    

output_t = tf.map_fn(window_func, tf.range(seqs),dtype=tf.float32)

output = sess.run(output_t)
```
